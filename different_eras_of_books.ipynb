**Calculation of geometric distance between graphs to check if there is a difference**



```python
dict1700_1800 = coding_for_project.dictOfLengthsMultiple(era1700_1800)
dict1800_1900 = coding_for_project.dictOfLengthsMultiple(era1800_1900)
dict1900_2000 = coding_for_project.dictOfLengthsMultiple(era1900_2000)

dict1700_1800,dict1800_1900,dict1900_2000
```

```python

def getDictDistance(dict1,dict2):
    distance = 0
    for key in dict1:
        if key < 10:
            distance += (dict1[key] + dict2[key])**2
    distance = distance ** 0.5
    return distance

getDictDistance(dict1700_1800,dict1800_1900),getDictDistance(dict1800_1900,dict1900_2000),getDictDistance(dict1700_1800,dict1900_2000)
# Can be seen that there is barely difference between 1900s and 2000s however whenever comparing to 1800s there is big difference. Maybe about renaissance and Age of Enlightenment "https://en.wikipedia.org/wiki/Age_of_Enlightenment"
```

**Zip Law**

**Find frequency of each word length**

**Order by power \- highest to lowest**

**Draw graph by log log of power and frequency**

**See if there are differences and if era gets identification and can be identified.**


```python
#Zip's Law
#https://www.ncbi.nlm.nih.gov/pmc/articles/PMC4176592/
```

```python
#dict1700_1800,dict1800_1900,dict1900_2000
```

```python

import math
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def drawLogGraphWithTrendLine(orderedList):
    rank = []
    freq = []
    for i in range(1, 10):
        rank.append(math.log(i))
        freq.append(math.log(orderedList[i-1]))
    plt.scatter(rank, freq)
    
    slope, intercept, _, _, _ = stats.linregress(rank, freq)
    plt.plot(np.array(rank), slope*np.array(rank) + intercept, color='red', label='Trend Line')
    
    plt.legend()
    plt.show()
    
    return slope
    
```

```python
coding_for_project.drawLogGraph(coding_for_project.orderDictByFrequency(dict1700_1800))
```

```python
print(coding_for_project.drawLogGraphWithTrendLine(coding_for_project.orderDictByFrequency(dict1700_1800)))
```

```python
coding_for_project.drawLogGraph(coding_for_project.orderDictByFrequency(dict1800_1900))
```

```python
print(coding_for_project.drawLogGraphWithTrendLine(coding_for_project.orderDictByFrequency(dict1800_1900)))
```
