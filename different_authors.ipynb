**Calculation of geometric distance between graphs to check if there is a difference**


```python
dict_author1 = coding_for_project.dictOfLengthsMultiple(author_1)
dict_author2 = coding_for_project.dictOfLengthsMultiple(author_2)
dict_author3 = coding_for_project.dictOfLengthsMultiple(author_3)
```

```python
def getDictDistance(dict1,dict2):
    distance = 0
    for key in dict1:
        if key < 13:
            distance += (dict1[key] + dict2[key])**2
    distance = distance ** 0.5
    return distance

getDictDistance(dict_author1,dict_author2),getDictDistance(dict_author2,dict_author3),getDictDistance(dict_author1,dict_author3)
```

```python
def getGradientsOfTrend(author):    
    for url in author:
        print(coding_for_project.getGradientOfTrendLine(coding_for_project.orderDictByFrequency(coding_for_project.dictOfLengths(url))))
    print("\n\n\n\n\n")
getGradientsOfTrend(author_1),getGradientsOfTrend(author_2),getGradientsOfTrend(author_3)
```

```python

def getRange(author):
    biggestGrad = coding_for_project.getGradientOfTrendLine(coding_for_project.orderDictByFrequency(coding_for_project.dictOfLengths(author[0])))
    smallestGrad = coding_for_project.getGradientOfTrendLine(coding_for_project.orderDictByFrequency(coding_for_project.dictOfLengths(author[0])))
    for url in author:
        grad = coding_for_project.getGradientOfTrendLine(coding_for_project.orderDictByFrequency(coding_for_project.dictOfLengths(url)))
        if grad < smallestGrad:
            smallestGrad = grad
        if grad > biggestGrad:
            biggestGrad = grad
    return abs(biggestGrad - smallestGrad)
getRange(author_1)
```

```python
def getAverage(author):
    sum = 0
    count = 0
    for url in author:
        count = count + 1
        sum = sum + coding_for_project.getGradientOfTrendLine(coding_for_project.orderDictByFrequency(coding_for_project.dictOfLengths(url)))
    return sum/count
getAverage(author_1)
```

```python
from tabulate import tabulate
col_names = ["Author Name","Range","Average","Range/Average"]
data = [["Shelley, Mary Wollstonecraft",getRange(author_1),getAverage(author_1),getRange(author_1)/getAverage(author_1)]
        ,["Austen, Jane",getRange(author_2),getAverage(author_2),getRange(author_2)/getAverage(author_2)]
        ,["Melville, Herman",getRange(author_3),getAverage(author_3),getRange(author_3)/getAverage(author_3)]]

print(tabulate(data, headers=col_names, tablefmt="fancy_grid"))
```
